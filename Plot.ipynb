```python
import torch
import numpy
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 20
x = [1000,2000,3000,4000,5000,6000,7000,8000,9000,10000]
#x = x.numpy()
#x = np.log(x)
e1 = torch.load('W1_10_50.pt',map_location=torch.device('cpu'))
e2 = torch.load('W2_10_50.pt',map_location=torch.device('cpu'))
e = e1 + e2
#e = np.log(e)
y = np.mean(e.numpy(),1)
#plt.plot(x, y, marker = 'o', label = 'd = 10')
error = np.std(e.numpy(),1)
plt.errorbar(x, y ,yerr = error, marker = 'o', label = 'd = 10', capsize=3, ecolor='tab:blue')

e1 = torch.load('W1_20_50.pt',map_location=torch.device('cpu'))
e2 = torch.load('W2_20_50.pt',map_location=torch.device('cpu'))
e = e1 + e2
#e = np.log(e)
y = np.mean(e.numpy(),1)
error = np.std(e.numpy(),1)
plt.errorbar(x, y ,yerr = error, marker = 'o', label = 'd = 20', capsize=3, ecolor='tab:orange')

plt.legend(loc='upper center', bbox_to_anchor=(0.5,1.2),ncol=2)
plt.xlabel("Number of training samples")
plt.ylabel("Translation loss")
plt.savefig('Traslation.pdf', bbox_inches='tight')
```
